```python
iris=pd.read_csv("iris.data",names=["a","b","c","d","e"])
```

```python
iris.head()
```

```python
iris.describe()
```

```python
import matplotlib.pyplot as plt
```

```python
#univariate - 1 col
plt.hist(iris.a)
#or iris["e"]
plt.title("histogram")
plt.xlabel("classes")
plt.ylabel("count")
plt.show()
```

```python
#bivariate - 2 col
plt.scatter(iris.a,iris.b)
plt.show()
```

```python
iris
```

```python
colors={"Iris-setosa":"r","iris-virginica":"g"}
fig, ax=plt.subplots() #merging many plots
size=len(iris.index)
print(size)
#plot each data plot
for i in range(size):
    ax.scatter(iris["a"][i],iris["b"][i],colors=colors[iris["e"][i]])
    
fig.show()
```

```python
iris.boxplot(column=["a","b","c","d"],rot=90)
#box displays 25% and 75%
```

```python
iris["e"].value_counts().plot.pie()
```

```python
iris["e"].value_counts()
```
